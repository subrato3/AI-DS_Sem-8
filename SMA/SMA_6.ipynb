```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Create an empty graph
social_network = nx.Graph()
num_nodes = 40
for i in range(num_nodes):
    social_network.add_node(i)
```

```python
probability_of_connection = 0.3

# Add edges (connections) between nodes based on the probability
for i in range(num_nodes):
    for j in range(i+1, num_nodes):
        if random.random() < probability_of_connection:
            social_network.add_edge(i, j)

nx.draw(social_network, with_labels=True)
plt.title("Random Social Network")
plt.show()
```

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(social_network)
print("Degree Centrality:", degree_centrality)
```

```python
# Calculate closeness centrality
closeness_centrality = nx.closeness_centrality(social_network)
print("Closeness Centrality:", closeness_centrality)
```

```python
# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(social_network)
print("Betweenness Centrality:", betweenness_centrality)
```

```python
# Identify most influential nodes based on centrality measures
most_influential_nodes_degree = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:5]
most_influential_nodes_closeness = sorted(closeness_centrality, key=closeness_centrality.get, reverse=True)[:5]
most_influential_nodes_betweenness = sorted(betweenness_centrality, key=betweenness_centrality.get, reverse=True)[:5]
print("Most Influential Nodes (Degree Centrality):", most_influential_nodes_degree)
print("Most Influential Nodes (Closeness Centrality):", most_influential_nodes_closeness)
print("Most Influential Nodes (Betweenness Centrality):", most_influential_nodes_betweenness)
```
